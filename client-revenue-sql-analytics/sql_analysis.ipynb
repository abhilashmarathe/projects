# **Step 1: Setup SQLite**

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("agency.db")
cursor = conn.cursor()
```

# **Step 2: Create Proper Relational Table**

```python
cursor.execute("""
CREATE TABLE clients (
    client_id INTEGER PRIMARY KEY,
    client_name TEXT,
    industry TEXT,
    city TEXT
)
""")

cursor.execute("""
CREATE TABLE projects (
    project_id INTEGER PRIMARY KEY,
    client_id INTEGER,
    project_type TEXT,
    project_cost INTEGER,
    start_date TEXT,
    payment_status TEXT,
    FOREIGN KEY (client_id) REFERENCES clients(client_id)
)
""")

conn.commit()
```

# **Step 4: Run Professional SQL Queries**

```python
pd.read_sql("SELECT SUM(project_cost) as total_revenue FROM projects", conn)
```

Revenue by City (JOIN)

```python
query = """
SELECT c.city, SUM(p.project_cost) as total_revenue
FROM clients c
JOIN projects p ON c.client_id = p.client_id
GROUP BY c.city
ORDER BY total_revenue DESC
"""
pd.read_sql(query, conn)
```

Revenue by Industry

```python
query = """
SELECT c.industry, COUNT(p.project_id) as total_projects,
SUM(p.project_cost) as total_revenue
FROM clients c
JOIN projects p ON c.client_id = p.client_id
GROUP BY c.industry
ORDER BY total_revenue DESC
"""
pd.read_sql(query, conn)
```

Advanced: Ranking Cities

```python
query = """
SELECT city, total_revenue,
RANK() OVER (ORDER BY total_revenue DESC) as revenue_rank
FROM (
    SELECT c.city, SUM(p.project_cost) as total_revenue
    FROM clients c
    JOIN projects p ON c.client_id = p.client_id
    GROUP BY c.city
)
"""
pd.read_sql(query, conn)
```
